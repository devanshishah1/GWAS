```python
import pandas as pd

data = pd.read_csv("/Users/devanshishah/Documents/GWAS1/data/data.tsv", sep='\t')

data.head()
```

```python
#different ways to "explore" the data

print("no of rows and columns:", data.shape)

if data.isnull().sum().any():
    print("null values:")
    print(data.isnull().sum())  # Show the count of missing values per column
else:
    print("data is clean, no null values")


#data.dtypes


print(data.describe())

print("unique values for each column:")
for column in data.columns:
    print(column, data[column].nunique())

```

```python
import pandas as pd
import numpy as np
import plotly.express as plot

data["-log10(p_value)"] = -np.log10(data['p_value'])

fig = plot.scatter(
    data,
    x='base_pair_location',  
    y='-log10(p_value)',    
    color='chromosome',      
    hover_data=['chromosome', 'base_pair_location', 'p_value'],  
    title='Interactive Manhattan Plot',
    labels={'base_pair_location': 'Base Pair Location', '-log10(p_value)': '-log10(p-value)'}
)

fig.show()
```
